```python
# Creating own datatype named as Fraction:-
class Fraction:

  # paramterised constructor -> jisko Input ki need hoti hai
  def __init__(self, x, y):
    self.numerator = x
    self.denominator = y

  def __str__(self):
    return '{}/{}'.format(self.numerator, self.denominator)

  def __add__(self,other):
    new_num,new_den = self.numerator*other.denominator + self.denominator*other.numerator, self.denominator*other.denominator
    return '{}/{}'.format(new_num,new_den)


  def __sub__(self,other):
    new_num,new_den = self.numerator*other.denominator - self.denominator*other.numerator, self.denominator*other.denominator
    return '{}/{}'.format(new_num,new_den)

  def __mul__(self,other):
    new_num,new_den = self.numerator*other.numerator, self.denominator*other.denominator
    return '{}/{}'.format(new_num,new_den)

  def __truediv__(self,other):
    new_num,new_den = self.numerator*other.denominator, self.denominator*other.numerator
    return '{}/{}'.format(new_num,new_den)


fr1 = Fraction(3,5)
fr2 = Fraction(1,3)
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```

```python
# Building 2D Geometry coordinates using OOPS Concepts:-
# 1. user can create and view 2d coordinate
# 2. user can find out the distance btw 2 coordinates
# 3. user can find out the distance of a coorinate from the origin
# 4. user can check if.a point lies on a. line
# 5. user can find out the distance btw a given 2d point &  a given line

# 1.
class Point:
  def __init__(self,x,y):
    self.x_cord = x
    self.y_cord = y

  def __str__(self):
    return '<{},{}>'.format(self.x_cord,self.y_cord)

# 2.
  def euclidean_distance(self,other):
    return ((self.x_cord - other.x_cord)**2 + (self.y_cord) - (other.y_cord)**2)**0.5

# 3.
  def distance_from_origin(self):
    # alternate method return (self.x_cord**2 + self.y_cord**2)**0.5
    return self.euclidean_distance(Point(0,0))   # no need to pass self again .self is an agrument that being passed in the 1st place , no need return self.euclidean_distance(self,Point(0,0)) -> throws an error

# 4. ax + by + c and user will tell a,b,c

  class Line:
    def __init__(self,a,b,c):
      self.a = a
      self.b = b
      self.c = c

    def __str__(self):
      return '{}x + {}y + {} = 0'.format(self.a,self.b,self.c)


    def point_on_line(line,point):
      if line.a*point.x_cord + line.b*point.y_cord + line.c == 0:
        return "lies on the line"

      else:
        return "does not lie on the Line"



# 5. user can find out the distance b/w a given 2D Point and a line
    def shortest_distance(line,point):
      return abs(line.a*point.x_cord + line.b*point.y_cord + line.c) / ((line.a**2 + line.b**2)**0.5)
```
