```python
# Guessing Problem
guess = int(input("Kindly Guess the Number"))
count = 1

while guess != jackpot:
    if guess < jackpot:
        print("Guess Higher")
    else:
        print("Guess Lower")
    guess = int(input("Kindly Guess the Number"))
    count+=1
else:
    print("Correct Guess")
    print("you took", count, "Attempts")
```
